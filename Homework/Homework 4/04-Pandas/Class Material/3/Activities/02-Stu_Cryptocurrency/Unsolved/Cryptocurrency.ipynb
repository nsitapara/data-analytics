```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
merged_table = pd.DataFrame.merge(bitcoin_df,dash_df, on="Date")
merged_table.head(5)
```

```python
# Rename columns so that they are differentiated

merged_table = merged_table.rename(columns={"Open_x":"Open_BTC","High_x":"High_BTC","Low_x":"Low_BTC","Close_x":"Close_BTC","Volume_x":"Volumne_BTC","Market Cap_x":"Market Cap_BTC","Open_y":"Open_Dash","High_y":"High_Dash","Low_y":"Low_Dash","Close_y":"Close_Dash","Volume_y":"Volumne_Dash","Market Cap_y":"Market Cap_Dash"

})
merged_table.head()
```

```python
# alternatively you can set your suffixes when the merge occurs

merged_table_suff = pd.DataFrame.merge(bitcoin_df,dash_df, on="Date",suffixes=(" Bitcoin"," Dash"))
merged_table_suff.head(5)
```

```python
# Collecting best open for Bitcoin and Dash

btcmaxopen = merged_table_suff["Open Bitcoin"].max()
dashmaxopen = merged_table_suff["Open Dash"].max()
# Collecting best close for Bitcoin and Dash
btcmaxclose = merged_table_suff["Close Bitcoin"].max()
Dashmaxclose = merged_table_suff["Close Dash"].max()

# Collecting the total volume for Bitcoin and Dash
BTC_volume = round(merged_table_suff["Volume Bitcoin"].sum()/1000000,1)
Dash_volume = round(merged_table_suff["Volume Dash"].sum()/1000000,1)
```

```python
# Creating a summary DataFrame using above values
summary_df = pd.DataFrame({"Best BTC Open": [btcmaxopen],
                           "Best BTC Close": [btcmaxclose],
                           "BTC Volumne": [BTC_volume],
                          "Best Dash Open": [dashmaxopen],
                            "Best Dash Close": [Dashmaxclose],
                           "Dash Volumn": [Dash_volume]})

summary_df.head()
```
