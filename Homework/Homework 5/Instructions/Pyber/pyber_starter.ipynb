```python

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data

city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# city_data.head()
# ride_data.head()
# Combine the data into a single dataset

city_rider_dataset = pd.merge(city_data,ride_data, how="outer")

city_rider_dataset
# Display the data table for preview
city_rider_dataset.columns
```

```python
city_rider_dataset = city_rider_dataset[["city","date","fare","ride_id","driver_count","type"]]
city_rider_dataset.head()
city_rider_dataset
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
# plt.scatter(x, y, s=z*1000, alpha=0.5)

city_rider_dataset


num_of_riders = city_rider_dataset["ride_id"]



# Build the scatter plots for each city types
grouped_by_type = city_rider_dataset.groupby("type")
grouped_by_type_count = grouped_by_type.count()
grouped_by_type_count 
grouped_by_type_average = grouped_by_type.mean()
grouped_by_type_average
# # # Create a legend

grouped_by_city = city_rider_dataset[["city","driver_count"]]
grouped_by_city = grouped_by_city.groupby("city")
grouped_by_city_num_riders_df = grouped_by_city.count()
grouped_by_city_num_riders_df



grouped_by_city_fare = city_rider_dataset[["city","fare"]]
grouped_by_city_fare = grouped_by_city_fare.groupby("city")
grouped_by_city_avg_fare = grouped_by_city_fare.mean()
grouped_by_city_avg_fare


grouped_by_city_type = city_rider_dataset[["city","driver_count","type"]]
grouped_by_city_type = grouped_by_city_type.groupby("type")
city_types = grouped_by_city_type.count()
city_types



# Incorporate a text label regarding circle size
plt.scatter(grouped_by_city_num_riders_df,grouped_by_city_avg_fare, edgecolors="black",s=grouped_by_city_num_riders_df*15,alpha=.5)

plt.xlabel("Total Number of Riders(Per City)")
plt.ylabel("Average Fare($)")
plt.show()
# plt.legend()
# Save Figure
```
